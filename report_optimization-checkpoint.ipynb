```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import os

df = read_csv('%s/optimization.log' % os.getcwd())
```

```python
df.head(5)
```

```python
ax = df\
    .groupby(['t'])['actor_count']\
    .mean()\
    .plot()
ax.set_ylabel('time')
ax.set_xlabel('number of actors')
```

```python
ax = df\
    .groupby(['t'])['T']\
    .mean()\
    .plot()
ax.set_ylabel('time')
ax.set_xlabel('tempreture')
```

```python
ax = df\
    .groupby(['actor_count'])['obj_f_value']\
    .mean()\
    .plot()
ax.set_ylabel('object value')
ax.set_xlabel('number of actors')
```

```python
ax = df\
    .groupby(['t'])['git_server_cpus']\
    .mean()\
    .plot()
ax.set_ylabel('cpu time')
ax.set_xlabel('time')
```

```python
ax = df\
    .groupby(['t'])['dE']\
    .mean()\
    .plot()
ax.set_ylabel('energy difference')
ax.set_xlabel('time')
```
